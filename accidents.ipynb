```python
import pandas as pd

accidents = pd.read_json('accidents.json')
```

```python
accidents.head()
```

```python
accidents.dtypes
```

```python
# Convert crash date to proper Pandas datetime

accidents['crash_date'] =  pd.to_datetime(accidents['crash_date'])
accidents.dtypes
```

```python
# Find Fatalities

fatalities = accidents.loc[(accidents['injuries_fatal'] > 0), :]
fatalities_clean = fatalities[['latitude', 'longitude']]
fatalities_clean.dropna(axis=0, inplace=True)
fatalities_clean.reset_index(drop=True)
```

```python
hit_n_run = accidents.loc[(accidents['hit_and_run_i'] == "Y"), :]
index = hit_n_run.index
number_of_rows = len(index)
number_of_rows
```

```python
fatalities_clean.to_json(r'fatalities.json')
```

```python
hit_n_run.to_json(r'hit_n_run.json')
```
