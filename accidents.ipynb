```python
import pandas as pd

accidents = pd.read_json('accidents.json')
```

```python
accidents.head()
```

```python
accidents.describe
```

```python
# Convert crash date to proper Pandas datetime
import datetime
accidents['crash_date'] =  pd.to_datetime(accidents['crash_date'])
accidents.dtypes
```

```python
# Find Fatalities

fatalities = accidents.loc[(accidents['injuries_fatal'] > 0), :]
fatalities_clean = fatalities[['latitude', 'longitude']]
fatalities_clean.dropna(axis=0, inplace=True)
fatalities_clean.reset_index(drop=True)
```

```python
fatalities_clean.to_json(r'fatalities.json')
```

```python
hit_n_run = accidents.loc[(accidents['hit_and_run_i'] == "Y"), :]
index = hit_n_run.index
number_of_rows = len(index)
number_of_rows
```

```python
hit_n_run_clean = hit_n_run[['crash_date','latitude', 'longitude']]
hit_n_run_clean.dropna(axis=0, inplace=True)
hit_n_run_clean.reset_index(drop=True)
```

```python
hit_n_run_clean.sort_values(by='crash_date')
```

```python
accidents_2020 = hit_n_run_clean.loc[(hit_n_run_clean['crash_date'] > '2019')].reset_index(drop=True)
accidents_2020
```

```python
accidents_2019 = hit_n_run_clean.loc[(hit_n_run_clean['crash_date'] < '2020') & (hit_n_run_clean['crash_date'] > '2018')].reset_index(drop=True)
accidents_2019
```

```python
accidents_2018 = hit_n_run_clean.loc[(hit_n_run_clean['crash_date'] < '2019') & (hit_n_run_clean['crash_date'] > '2017')].reset_index(drop=True)
accidents_2018
```

```python
accidents_2017 = hit_n_run_clean.loc[(hit_n_run_clean['crash_date'] < '2018') & (hit_n_run_clean['crash_date'] > '2016')].reset_index(drop=True)
accidents_2017
```

```python
accidents_2020.to_json(r'hit_n_run_2020.json')
accidents_2019.to_json(r'hit_n_run_2019.json')
accidents_2018.to_json(r'hit_n_run_2018.json')
accidents_2017.to_json(r'hit_n_run_2017.json')
```
